```python
input = None
with open("input.txt", 'r') as f:
    temp_input = [l.replace('\n', '') for l in f.readlines()]
    input = [[c for c in l] for l in temp_input]
input 
```

```python
import numpy as np

array = np.array(input)
chars = np.unique(array)
antennae_types = chars[np.where(chars != '.')]
print(antennae_types)
antennae_map = {}
antennae_coords = [(int(c[0]), int(c[1])) for c in np.dstack(np.where(array != '.')).squeeze(0)]
print(antennae_coords)
for a in antennae_types:
    antennae_map[a] = np.dstack(np.where(array == a)).squeeze(0)
antennae_map
```

```python
antinodes = set()
for a in antennae_types:
    coords = antennae_map[a]
    for (i, cx) in enumerate(coords):
        cys = np.delete(coords, i, axis=0)
        for cy in cys:
            dist = cy - cx
            mults = 0
            while True:
                new_loc = (cy + dist * mults) 
                candidate_new_loc = (int(new_loc[0]), int(new_loc[1]))
                if (new_loc[0] >= 0 and new_loc[0] < array.shape[0]) and (new_loc[1] >= 0 and new_loc[1] < array.shape[1]):
                    antinodes.add(candidate_new_loc)
                    mults += 1
                else:
                    break
print(antinodes)        
print(len(antinodes))
```

```python
for an in antinodes:
    array[an[0], an[1]] = '#'
print(array)
```
