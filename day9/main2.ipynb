```python
import numpy as np # type: ignore
import sys
np.set_printoptions(threshold=sys.maxsize)
```

```python
input = None
with open("input.txt", 'r') as f:
    input = [int(i) for i in f.read()]
input
```

```python
is_free = False
file_index, index_tracker = 0, 0
array = np.array([-1] * sum(input))
for n in input:
    if is_free:
        index_tracker += n
        is_free = False
    else:
        array[index_tracker:index_tracker+n] = file_index
        index_tracker += n
        file_index += 1
        is_free = True
array
```

```python
mut_array = array.copy()
file_map = []
free_map = {}

curr_block = []
for i in range(len(array)):

    if i > 0 and i < len(array) - 1:
        if array[i] == array[i-1]:
            curr_block.append(i)
        else:
            if array[i-1] == -1:
                if len(curr_block) not in free_map.keys():
                    free_map[len(curr_block)] = [curr_block.copy()]
                else:
                    free_map[len(curr_block)].insert(0, curr_block.copy())
            else:
                file_map.append(curr_block)
            curr_block = [i]

    if i == 0:
        curr_block.append(i)

    if i == len(array) - 1:
        if array[i] == array[i-1]:
            curr_block.append(i)
            if array[i-1] == -1:
                if len(curr_block) not in free_map.keys():
                    free_map[len(curr_block)] = [curr_block.copy()]
                else:
                    free_map[len(curr_block)].insert(0, curr_block.copy())
            else:
                file_map.append(curr_block)
        else:
            if array[i-1] == -1:
                if len(curr_block) not in free_map.keys():
                    free_map[len(curr_block)] = [curr_block.copy()]
                else:
                    free_map[len(curr_block)].insert(0, curr_block.copy())
            else:
                file_map.append(curr_block)
            # handle current element
            if array[i] == -1:
                if 1 not in free_map.keys():
                    free_map[1] = [[i]]
                else:
                    free_map[1].insert(0, [i])
        

print('files:', file_map)
print('free spaces:', free_map)
```

```python
free_space_list = []
for i in range(len(array)):
    # if i > 0 and i < len(array) - 1:
    if array[i-1] == -1 and array[i] != -1:
        f = len(file_map) - 1
        while f >= 0:
            free_space_length = len(free_space_list)
            if free_space_length == 0:
                break
            if free_space_length >= len(file_map[f]) and not free_space_list[0] >  file_map[f][0]:
                file_indexes = file_map.pop(f)
                # print(free_space_list)
                mut_array[free_space_list[0]:free_space_list[0] + len(file_indexes)] = mut_array[file_indexes]
                mut_array[file_indexes] = -1
                free_space_list = free_space_list[len(file_indexes):]
                # print(mut_array)
                if free_space_length == 0:
                    break
            f -= 1
        free_space_list = []
    elif array[i] == -1:
        free_space_list.append(i)
```
